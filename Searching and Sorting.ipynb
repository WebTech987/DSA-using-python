# Quick Sort
Quick Sort is a Divide and Conquer algorithm. It picks an element as a pivot and partitions the given array around the picked pivot.
Given an array arr[], its starting position is low (the index of the array) and its ending position is high(the index of the array).

Note: The low and high are inclusive.
Implement the partition() and quickSort() functions to sort the array.


Approach: We are trying to fit one element at a time on its correct position and that's pivot element
steps:
1. Define pivot element-index
2. Iterate in the rest of the array from left and right both sides
3. if element less than the pivot is found in left operation then moves to next element
    else if the element is greater than the pivot stop here
4. in the right portion skips the elements greater than equal to pivot, else 
    if element in right portion less than pivot element is found stop

5. Now swap the greater element of less portion with the greater element of right portion

6. keep doing the same until the left pointer and right pointer meets and a point(correct position of pivot)

7. swap the pivot element and the element at the position where pointers meet


----here we get two parts left of the pivot and right of the pivot ----
Do the same for both parts recursively


![image.gif](https://www.tutorialspoint.com/data_structures_algorithms/images/quick_sort_partition_animation.gif)

```python
class Solution:
    # Function to sort a list using the Quick Sort algorithm.
    def quickSort(self, arr, low, high):
        if low < high:
            # Partition the array and get the index of the pivot element.
            j = self.partition(arr, low, high)
            # Recursively sort the subarrays on both sides of the pivot.
            Solution.quickSort(self, arr, low, j - 1)
            Solution.quickSort(self, arr, j + 1, high)
        
        return arr
    
    # Function to partition the array and return the index of the pivot element.
    def partition(self, arr, low, high):
        pivot = arr[low]  # Choose the pivot element (first element in this case).
        i = low + 1
        j = high
        
        while i <= j:
            while i <= j and arr[i] <= pivot:
                i += 1
            while i <= j and arr[j] >= pivot:
                j -= 1
            if i <= j:
                arr[i], arr[j] = arr[j], arr[i]  # Swap elements to maintain the partition.
        
        arr[low], arr[j] = arr[j], arr[low]  # Move the pivot element to its correct position.
        return j  # Return the index of the pivot element.
```

#  Heap Sort:


```python
class Solution:
    # Heapify function to maintain heap property.
    def heapify(self, arr, n, i):
        # 'n' is the size of the heap, and 'i' is the index of the current node.
        
        # Base case: If the heap has only one element or is empty, return.
        if n <= 1:
            return
        
        largest = i  # Initialize the largest as the root node.
        
        l = 2 * i + 1  # Calculate the left child index.
        r = 2 * i + 2  # Calculate the right child index.
        
        # Compare the left child with the current largest.
        if l < n and arr[l] > arr[largest]:
            largest = l
        
        # Compare the right child with the current largest.
        if r < n and arr[r] > arr[largest]:
            largest = r
        
        if largest != i:
            # If the largest is not the current node, swap them.
            arr[i], arr[largest] = arr[largest], arr[i]
            
            # Recursively heapify the affected subtree.
            self.heapify(arr, n, largest)
    
    # Function to build a Heap from an array.
    def buildHeap(self, arr, n):
        # 'n' is the size of the array.
        
        # Start from the last non-leaf node (n//2 - 1) and move to the root.
        st_idx = n // 2 - 1
        
        for i in range(st_idx, -1, -1):
            # Heapify each subtree to ensure the heap property.
            self.heapify(arr, n, i)
    
    # Function to sort an array using Heap Sort.
    def HeapSort(self, arr, n):
        # 'n' is the size of the array.
        
        # Build a max-heap from the given array.
        self.buildHeap(arr, n)

        i = n - 1  # Initialize the last index.
        
        # Extract elements one by one from the heap and maintain the heap property.
        while i > 0:
            # Swap the root (maximum element) with the last element.
            arr[0], arr[i] = arr[i], arr[0]
            
            # Reduce the size of the heap and heapify the root.
            self.heapify(arr, i, 0)
            
            i -= 1  # Move to the next index.
        
        return arr  # Return the sorted array.
```

# Selection Sort:

```python
class Solution:
    def selectionSort(self, arr, n):
        # Iterate through the entire array.
        for i in range(n):
            # Find the minimum element in the remaining unsorted portion.
            min_index = i
            for j in range(i + 1, n):
                if arr[j] < arr[min_index]:
                    min_index = j

            # Swap the minimum element with the current element (arr[i]).
            arr[i], arr[min_index] = arr[min_index], arr[i]

        return arr
```

# Find missing and repeating

Given an unsorted array Arr of size N of positive integers. One number 'A' from set {1, 2,....,N} is missing and one number 'B' occurs twice in array. Find these two numbers.

```python
# User function Template for python3
class Solution:
    def findTwoElement(self, arr, n):
        # arr: The input list containing positive integers from 1 to N with one missing number and one duplicate number.
        # n: The size of the input list.

        # Sort the array to bring the numbers in ascending order, as the missing and duplicate numbers are likely to be nearby.
        arr.sort()

        # Initialize variables to store the missing and duplicate numbers.
        missing = None
        repeat = None

        # Loop through the array to find the first missing number.
        for i in range(n):
            if i + 1 != arr[i]:
                missing = i + 1
        
        # Check if the first missing number is also present in the array.
        # If it is, then there is a duplicate number that matches the first missing number,
        # so we need to find the second missing number.
        if missing in arr:
            for i in range(n):
                if i + 1 != arr[i]:
                    missing = i + 1
                    break

        # Loop through the sorted array to find the duplicate number.
        # We check for adjacent elements, as they are likely to be the duplicate and the missing number.
        for i in range(1, n):
            if arr[i] == arr[i - 1]:
                repeat = arr[i]
                break

        # If the missing and duplicate numbers are the same, it means the missing number is the maximum value.
        # We need to adjust the missing number to be one less than the maximum value.
        if missing == repeat:
            # Adjust the missing number to be one less than the maximum value (N) as the array size is (N-1).
            missing = missing - 1

        # Return the duplicate and missing numbers.
        return repeat, missing
```

# Finding the majority element in an array
Given an array A of N elements. Find the majority element in the array. A majority element in an array A of size N is an element that appears more than N/2 times in the array.

```python
# Given an array A of N elements. Find the majority element in the array.
# A majority element in an array A of size N is an element that appears more than N/2 times in the array.

class Solution:
    def majorityElement(self, A, N):
        # A: The input list containing elements.
        # N: The number of elements in the input list.

        # Step 1: Sort the array to bring identical elements together.
        A.sort()
        n = N
        count = 1
        
        # Step 2: Handle the special case when there is only one element in the array.
        # If there's only one element, it is the majority element.
        if n == 1:
            return A[0]
        
        # Step 3: Loop through the sorted array to find the majority element.
        # The majority element will always be at the middle position in the sorted array (index N//2).
        # This is because the majority element appears more than N/2 times.
        for i in range(1, n):
            if A[i] == A[i-1]:
                count += 1
                
            # Check if the current element's count is greater than N/2 (the majority condition).
            # If it is, then the current element is the majority element.
            if count > N//2:
                return A[i]
            
            # If the current element is different from the previous element,
            # reset the count to start counting the next element.
            elif A[i] != A[i-1]:
                count = 1

        # If there is no majority element, return -1 as per the problem statement.
        return -1
```

# Find Pair Given Difference

Given an array Arr[] of size L and a number N, you need to write a program to find if there exists a pair of elements in the array whose difference is N.

```python
class Solution:
    def findPair(self, arr, L, N):
        # arr: The input list containing elements.
        # L: The number of elements in the input list.
        # N: The target difference.

        # Step 1: Sort the array to bring elements in ascending order.
        arr.sort()

        # Initialize two pointers i and j, pointing to the first and second elements, respectively.
        i = 0
        j = 1

        # Step 2: Use two-pointer approach to find the pair of elements with the required difference N.
        while i < L and j < L:
            # Check if the current elements at positions i and j satisfy the required difference condition.
            if i != j and abs(arr[i] - arr[j]) == N:
                return True
            # If the difference is less than N, increment the j pointer to consider a larger difference.
            elif abs(arr[i] - arr[j]) < N:
                j += 1
            # If the difference is greater than N, increment the i pointer to consider a smaller difference.
            else:
                i += 1

        # If no pair with the required difference is found, return False.
        return False
```

# Find All Four Sum Numbers
Given an array of integers and another number. Find all the unique quadruple from the given array that sums up to the given number.

```python
class Solution:
    def fourSum(self, nums, target):
        # nums: The input list containing integers.
        # target: The target sum for the quadruples.

        # Sort the input array to help in handling duplicates and to enable the two-pointer approach.
        nums.sort()
        n = len(nums)
        ans = []

        # First loop to traverse the array and consider the first element of the quadruple.
        for i in range(n):
            # Skip duplicates while moving i. This ensures that each element is considered only once as the first element.
            if i != 0 and nums[i] == nums[i-1]:
                continue

            # Second loop to traverse the array and consider the second element of the quadruple.
            for j in range(i+1, n):
                # Skip duplicates while moving j. This ensures that each element is considered only once as the second element.
                if j != i+1 and nums[j] == nums[j-1]:
                    continue

                # Two pointers, k and l, are used to find the other two elements of the quadruple.
                k = j + 1
                l = n - 1

                while k < l:
                    # Calculate the sum of the four elements (nums[i], nums[j], nums[k], nums[l]).
                    s = nums[i] + nums[j] + nums[k] + nums[l]

                    if s > target:
                        # If the sum is greater than the target, move the right pointer (l) to consider a smaller value.
                        l -= 1
                    elif s < target:
                        # If the sum is less than the target, move the left pointer (k) to consider a larger value.
                        k += 1
                    else:
                        # If the sum is equal to the target, add the quadruple to the result list (ans).
                        ans.append([nums[i], nums[j], nums[k], nums[l]])

                        # Skip duplicates while moving k and l.
                        while k < l and nums[k] == nums[k+1]:
                            k += 1
                        while k < l and nums[l] == nums[l-1]:
                            l -= 1

                        # Move both pointers to find new elements for the quadruple.
                        k += 1
                        l -= 1

        return ans
```

# Count triplets with sum smaller than X
Given an array arr[] of distinct integers of size N and a value sum, the task is to find the count of triplets (i, j, k), having (i<j<k) with the sum of (arr[i] + arr[j] + arr[k]) smaller than the given value sum.



```python
class Solution:
    def countTriplets(self, arr, n, sumo):
        arr.sort()  # Sort the input array in ascending order
        
        count = 0  # Initialize a variable to keep track of the count of valid triplets
        
        for i in range(n - 2):  # Iterate through the array up to the third-to-last element
            j = i + 1  # Initialize a pointer j starting from the next element of i
            k = n - 1  # Initialize a pointer k starting from the last element
            
            while j < k:  # While j and k pointers don't cross each other
                if arr[i] + arr[j] + arr[k] < sumo:  # If the sum of elements is less than the target sum
                    # Explanation: If the sum of arr[i], arr[j], and arr[k] is less than the target sum,
                    # then all the elements between j and k (inclusive) will also form valid triplets.
                    # Therefore, we add the count of valid triplets between j and k to the 'count' variable.
                    count += k - j
                    j += 1  # Increment the j pointer to consider the next element
                else:
                    k -= 1  # If the sum is not less than the target, decrement the k pointer
                    
        return count  # Return the total count of valid triplets
```

# Subarrays with sum equal to Zero
You are given an array arr[] of size n. Find the total count of sub-arrays having their sum equal to 0.



```python
class Solution:
    def findSubArrays(self, arr, n):
        map1 = {}  # A dictionary to store the frequency of prefix sums
        
        presum = 0  # Initialize prefix sum variable to 0
        count = 0   # Initialize count variable for subarrays with sum 0
        
        map1[presum] = 1  # Initialize the prefix sum dictionary with 0 frequency
        
        for i in range(n):  # Loop through the array
            presum += arr[i]  # Update the prefix sum
            
            if presum in map1:  # Check if the current prefix sum exists in the dictionary
                count += map1[presum]  # If it does, increment the count by the frequency of that prefix sum
                map1[presum] += 1  # Increment the frequency of the prefix sum in the dictionary
            else:
                map1[presum] = 1  # If the prefix sum doesn't exist, initialize it with frequency 1
        
        return count  # Return the total count of subarrays with sum 0
```

# Merging two unsorted arrays in sorted order
GIven two different arrays A and B of size N and M, the task is to merge the two arrays which are unsorted into another array which is sorted.

```python
class Solution:
    def sortedMerge(self, a, b, c, n, m):
        # Merging two unsorted arrays and sorting them
        ans = self.mergesortArr(a + b)
        
        # Copying the sorted merged array to the output array c
        c[:] = ans
        
        return c
        
    
    def mergesortArr(self, arr):
        n = len(arr)
        
        # Base cases for sorting
        if len(arr) == 1:
            return arr  # Array of length 1 is already sorted
        elif len(arr) == 2:
            if arr[0] >= arr[1]:
                return arr[::-1]  # Swap the elements if needed to sort
            return arr
        
        else:
            # Recursive sorting of the first and second halves
            a = self.mergesortArr(arr[:n//2])  # Sort the first half
            b = self.mergesortArr(arr[n//2:])  # Sort the second half
            
            n1 = len(a)
            m1 = len(b)
            
            ans = []  # Initialize the merged array
            i = 0
            j = 0
            while i < n1 and j < m1:
                if a[i] <= b[j]:
                    ans += [a[i]]  # Add the smaller element to ans from the first half
                    i += 1
                else:
                    ans += [b[j]]  # Add the smaller element to ans from the second half
                    j += 1
                
            # Add remaining elements from a and b
            ans += a[i:]  # Add remaining elements from the first half
            ans += b[j:]  # Add remaining elements from the second half
            
            return ans  # Return the merged and sorted array
```

# Merge Without Extra Space

```python
# TLE Approach

class Solution:
    
    # Function to merge the arrays.
    def merge(self, arr1, arr2, n, m):
        # Compare the last element of arr1 with the first element of arr2
        if arr1[-1] < arr2[0]:
            return arr1, arr2  # No merging needed if arr1's largest element is smaller than arr2's smallest element
        
        else:
            i = 0  # Initialize pointer i to traverse arr1
            
            while i < n:
                x = i  # Store the current position of i
                
                j = 0  # Initialize pointer j to traverse arr2
                
                # Compare elements of arr1 with elements of arr2
                while j < m:
                    i = x  # Reset i to the stored position
                    
                    # Check if the current element of arr1 is greater than the current element of arr2
                    if j < m and arr1[i] > arr2[j]:
                        # Swap the elements of arr1 and arr2
                        arr1[i], arr2[j] = arr2[j], arr1[i]
                        
                        # Move the swapped element in arr1 to its correct sorted position
                        while i > 0:
                            if arr1[i] < arr1[i - 1]:
                                arr1[i], arr1[i - 1] = arr1[i - 1], arr1[i]
                                i -= 1
                            else:
                                break
                        
                        xx = j  # Store the current position of j
                        
                        # Move the swapped element in arr2 to its correct sorted position
                        while j < m - 1:
                            if arr2[j + 1] < arr2[j]:
                                arr2[j + 1], arr2[j] = arr2[j], arr2[j + 1]
                                j += 1
                            else:
                                break
                        
                        j = xx  # Reset j to the stored position
                        
                        break  # Exit the loop
                        
                    else:
                        j += 1  # Move to the next element of arr2
                
                i += 1  # Move to the next element of arr1
            
            return arr1, arr2
```

# Merge K sorted/unsorted arrays

```python
#TLE Approach
class Solution:
    #Function to merge k sorted arrays.
    def mergeKArrays(self, arr, K):
        # code here
        # return merged list
        ans = arr[0]
        
        for i in range(1, K):
            for j in range(K):
                
                ans += [arr[i][j]]
                
                p = len(ans)-1
                
                while p > 0 and ans[p] < ans[p-1] :
                    ans[p], ans[p-1] = ans[p-1], ans[p]
                    
                    p -= 1

        return ans
```

# Minimum Swaps to Sort
Given an array of n distinct elements. Find the minimum number of swaps required to sort the array in strictly increasing order.

```python
class Solution:
    # Function to find the minimum number of swaps required to sort the array.
    def minSwaps(self, nums):
        n = len(nums)  # Get the length of the array
        
        #We will go in reverse order for the sorted array we will tranform in the the form of given array and we will count
        
        # Create pairs of element and their indices for sorting
        for i in range(n):
            nums[i] = [nums[i], i]
        
        nums.sort()  # Sort the array of pairs
        
        i = 0
        count = 0  # Initialize the count of swaps
        
        # Iterate through the array to count the swaps required
        
        while i < n:
            # i is incremented only when the element gets on it's proper position
            if nums[i][1] == i:
                i += 1  # Move to the next element
                
            # else the keep swapping till the correct position is achieved    
            else:
                count += 1  # Increment the swap count
                # Swap the elements to bring them to their correct positions
                
                nums[nums[i][1]], nums[i] = nums[i], nums[nums[i][1]]
        
        return count  # Return the minimum number of swaps
```

# Allocate minimum number of pages
You have N books, each with Ai number of pages. M students need to be allocated contiguous books, with each student getting at least one book. Out of all the permutations, the goal is to find the permutation where the student with the most books allocated to him gets the minimum number of pages, out of all possible permutations.

Note: Return -1 if a valid assignment is not possible, and allotment should be in contiguous order (see the explanation for better understanding).

 



```python
class Solution:
    # Function to find minimum number of pages.
    def findPages(self, A, N, M):
        if M > N:
            return -1
        elif M == N:
            return max(A)  # The student with most books gets the maximum pages

        min1 = 999999
        l = max(A)  # Start with the minimum possible allocation (max pages)
        h = sum(A)  # End with the maximum possible allocation (all pages)

        # Binary search to find the optimal allocation
        while l <= h:
            mid = (l + h) // 2
            if self.solveit(A, N, M, mid):
                min1 = min(min1, mid)
                h = mid - 1  # Try to minimize the allocation
            else:
                l = mid + 1   # Increase the allocation

        return min1

    def solveit(self, A, N, M, mid):
        count = 1
        n = N
        sum1 = 0

        for i in range(n):
            if sum1 + A[i] <= mid:
                sum1 += A[i]
            else:
                count += 1
                if count > M or A[i] > mid:
                    return False
                sum1 = A[i]  # Reset sum1 for the new student allocation

        return True
```

# Count Inversions

Given an array of integers. Find the Inversion Count in the array. 

Inversion Count: For an array, inversion count indicates how far (or close) the array is from being sorted. If array is already sorted then the inversion count is 0. If an array is sorted in the reverse order then the inversion count is the maximum. 
Formally, two elements a[i] and a[j] form an inversion if a[i] > a[j] and i < j.
 

```python
# TLE Approach

class Solution:
    #User function Template for python3
    
    # arr[]: Input Array
    # N : Size of the Array arr[]
    #Function to count inversions in the array.
    def inversionCount(self, arr, n):
        
        ans = 0
    
        for i in range(n):
            
            j = 0 
            count = 0
            
            while j < i :
                
                if arr[i] < arr[j] :
                    
                    count += 1
                j += 1
            ans += count
        
        return ans
```

```python
# Optimized approach
class Solution:
    # User function to count inversions in the array.
    def inversionCount(self, arr, n):
        # Initialize the inversion count
        ans = 0
        
        # Base case: If the array size is 0 or 1, there are no inversions
        if n <= 1:
            return 0

        # Recursive function to perform merge sort and count inversions
        def mergesort(Arr):
            nonlocal ans
            
            # Get the length of the array
            m = len(Arr)
            
            # Base case: If the array size is 1, return the array
            if m == 1:
                return Arr
            
            # Recursively split the array into halves and sort them
            a, b = mergesort(Arr[: m // 2]), mergesort(Arr[m // 2:])
            
            # Initialize pointers for the left and right halves
            l = 0
            r = 0
            
            # Merge the sorted halves while counting inversions
            while l < len(a) and r < len(b):
                if a[l] <= b[r]:
                    l += 1
                else:
                    # Count inversions when a[l] > b[r]
                    ans += len(a) - l
                    r += 1
            
            # Return the sorted merged array
            return sorted(a + b)
        
        # Call the mergesort function and return the inversion count
        mergesort(arr)
        return ans
```

# Minimum element in a sorted and rotated array

A sorted(in ascending order) array A[ ] with distinct elements is rotated at some unknown point, the task is to find the minimum element in it.

```python
class Solution:
    def findMin(self, arr, n):
        # Initialize the left (l) and right (h) pointers
        l = 0
        h = n - 1
        
        # Binary search to find the minimum element
        while l <= h:
            # Calculate the mid index
            mid = (l + h) // 2
            
            # Check if mid is the last element before rotation
            # and the next element is greater than the last element
            if mid == h - 1 and arr[mid] > arr[h]:
                return arr[h]
            
            # If the mid element is greater than the last element,
            # the minimum element is on the right side of mid
            if arr[mid] > arr[h]:
                l = mid + 1
            
            # If the mid element is less than the last element,
            # the minimum element is on the left side of mid or is mid itself
            elif arr[mid] < arr[h]:
                h = mid
                
        # If the loop ends without returning, the minimum element is at the start
        return arr[0]
```

# Search in Rotated Array 2
Given a sorted and rotated array A of N elements which is rotated at some point, and may contain duplicates and given an element key. Check whether the key exist in the array A or not.

```python
class Solution:
    def Search(self, n, arr, k):
        # If the array contains only one element and it matches the key
        if n == 1 and arr[0] == k:
            return 1
        
        # Initialize the left (l) and right (h) pointers
        l = 0
        h = n - 1
        
        # Binary search to check if the key exists in the rotated array
        while l <= h:
            # Calculate the middle index mid
            mid = (l + h) // 2
            
            # Check if the middle element, left element, or right element is the key
            if arr[mid] == k or arr[l] == k or arr[h] == k:
                return 1  # Key found
                
            # If the middle element is greater than the key, move l towards mid
            if arr[mid] > k:
                l = mid + 1
            
            # If the middle element is less than the key, move h towards mid
            elif arr[mid] < k:
                h = mid - 1
        
        return 0  # Key not found

```

# Kth element of two sorted arrays
Given two sorted arrays arr1 and arr2 of size N and M respectively and an element K. The task is to find the element that would be at the kth position of the final sorted array.

```python
class Solution:
    def kthElement(self, arr1, arr2, n, m, k):
        i = 0
        j = 0
        
        # If k is 1, return the minimum of the first elements of both arrays
        if k == 1:
            return min(arr1[0], arr2[0])
            
        count = 0
        
        # Merge the two sorted arrays to find the kth element
        while i < n and j < m:
            # Compare the elements of arr1 and arr2
            if arr1[i] <= arr2[j]:
                i += 1
                count += 1
                
                # If kth element is found, return it
                if count == k:
                    return arr1[i - 1]
            
            elif arr1[i] > arr2[j]:
                j += 1
                count += 1
                
                # If kth element is found, return it
                if count == k:
                    return arr2[j - 1]
        
        # If kth element is not found in the merged part of arr1 and arr2
        # Search for the kth element in the remaining part of arr1 and arr2
        while i < n:
            count += 1
            if count == k:
                return arr1[i]
            i += 1
        
        while j < m:
            count += 1
            if count == k:
                return arr2[j]
            j += 1
        
        return -1  # Return -1 if k is out of bounds
```

Explanation:
The integers in their binary

hence the non-increasing sorted order is:
{15}, {7}, {5, 3, 9, 6}, {2, 4, 32}

```python
#TLE approach

from collections import defaultdict

class Solution:
    def sortBySetBitCount(self, arr, n):
        # Your code goes here
        no = ''

        target_dict = {}
        max1 = 0
        
        # Calculate the set bit count for each number and store it in target_dict
        for i in range(n):
            no = 0
            count = self.DecimalToBinary(arr[i], no)
            
            target_dict[arr[i]] = count
            max1 = max(max1, count)
        
        ans = []
        
        # Iterate over the set bit counts in decreasing order
        for max1 in range(max1, -1, -1):
            # Collect elements with the current set bit count and append to ans
            for i in range(n):
                if target_dict[arr[i]] == max1:
                    ans.append(arr[i])
            
            max1 -= 1
            
        # Update the input array with the sorted elements
        arr[:] = ans
        return arr

    def DecimalToBinary(self, num, no):
        # Convert a decimal number to binary and count set bits
        if num >= 1:
            no = self.DecimalToBinary(num // 2, no)
            if str(num % 2) == '1':
                no += 1 
        return no
```

```python


# Function to calculate the total time needed to cook 'p' pratas by 'l' cooks with given ranks 'arr'
def totalTime(p, l, arr):
    # If only 1 prata is required, return the minimum time taken by any cook
    if p == 1:
        return min(arr)
        
    x = max(arr)  # Find the maximum rank among the cooks
    
    l = 0
    h = (x * (p) * (p + 1)) // 2  # Calculate the upper bound using the sum of first 'p' terms of an arithmetic series
    ans = 0  # Variable to store the final answer
    
    # Perform binary search to find the minimum time required
    while l < h:
        mid = (l + h) // 2  # Calculate the mid time
        
        # Check if it's possible to cook 'p' pratas within the time 'mid'
        if noparats(p, arr, mid):
            ans = mid  # Update the answer if the condition is met
            h = mid - 1  # Move the upper bound to the left
        else:
            l = mid + 1  # Move the lower bound to the right
    
    return ans  # Return the final answer


# Function to check if it's possible to cook 'p' pratas within the time 't' using the given cooks
def noparats(p, arr, t):
    n = len(arr)  # Number of cooks
    count = 0  # Counter to keep track of pratas cooked
    
    # Iterate through each cook's rank
    for i in range(n):
        time = arr[i]  # Time taken by the current cook to cook 1 prata
        j = 2  # Initialize j to 2
        
        # Iterate while it's possible for the current cook to cook more pratas within the time 't'
        while time <= t:
            count += 1  # Increment the counter
            time += arr[i] * j  # Calculate the time for the next prata
            j += 1  # Increment j
            
    # If the total pratas cooked are greater than or equal to 'p', return True
    if count >= p:
        return True
    else:
        return False  # Otherwise, return False
```

```python
# Unoptimized Approach
def largestsum(arr1, arr2):
    n1 = len(arr1)
    n2 = len(arr2)
    cmn = [[0, 0]]

    i = 0
    j = 0
    
    while i < n1 and j < n2 :
        
        if arr1[i] == arr2[j] :
            cmn += [[i, j]]
            i += 1
            j += 1
        
        elif arr1[i] < arr2[j] :
            i += 1
            
        else:
            j += 1
    
    cmn += [[n1, n2]]
    
    n = len(cmn)
    
    maxsum = 0
    
    for i in range(n-1):
        
        sum1 = sum(arr1[cmn[i][0]  : cmn[i+1][0]])
        sum2 = sum(arr2[cmn[i][1]  : cmn[i+1][1]])
        
        maxsum += max(sum1, sum2)
        
    return maxsum


arr1 = [3, 5, 7, 9, 20, 25, 30, 40, 55, 56, 57, 60, 62]
arr2 = [1, 4, 7, 11, 14, 25, 44, 47, 55, 57, 100]
print(largestsum(arr1, arr2))

arr3 = [-5, 100, 1000, 1005]
arr4 = [-12, 1000, 1001]
print(largestsum(arr3, arr4))
```
